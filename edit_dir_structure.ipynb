```python
logs = glob.glob("../results/**/*.txt", recursive=True)
```

```python
model2dir = {"trans-bpe": "trans", "lstm-small": "lstm", "rnn-small": "rnn", "trans-bpe-action": "trans", "lstm-small-action": "lstm", "rnn-small-action": "rnn"}
for log in logs:
    lang = log.split("/")[2]
    fold = log.split("/")[3].replace("fold_","")
    model = log.split("/")[5]
    if "action" in model:
        action = log.split("/")[6] + "/"
        if action not in ["td/", "lc-as/"]:
            continue
    else:
        action = ""
    if model not in model2dir:
        continue
    model_dir = model2dir[model]
    new_path = f"work/results/{lang}/{fold}/{action}{model_dir}"
    try:
        shutil.move(log, new_path)
    except:
        pass
```
